## Filtering
I realize that some of these could be done in a single line but I'm breaking them up for readability.

```python
#List of free races
free = [i.title() for i in ['AARAKOCRA', 'CHANGELING', 'DRAGONBORN', 'DWARF', 'ELF', 'GENASI',
                            'GNOME', 'GOLIATH', 'HALF-ELF', 'HALF-ORC', 'HALFLING', 'HUMAN',
                            'KALASHTAR', 'SHIFTER', 'TIEFLING', 'AASIMAR', 'WARFORGED']]
```

```python
free
```

### Setting up DFs
- Grouped by Race
- separated or notated on 'free'
- adds percentage column

```python
all_races = dndf.groupby('Race').sum().sort_values(by='Count')
```

```python
total_count = all_races.sum()[0]
```

```python
all_races['Percentage'] = [((i / total_count) * 100).round(2) for i in all_races['Count']]
```

```python
all_races = all_races.reset_index()
```

```python
all_races['Free'] = all_races['Race'].isin(free)
```

```python
free_races = dndf[dndf['Race'].isin(free)].groupby('Race').sum().sort_values(by='Count')
```

```python
total_free = free_races.sum()[0]
```

```python
free_races['Percentage'] = [((i / total_free) * 100).round(2) for i in free_races['Count']]
```

```python
free_races = free_races.reset_index()
```

```python
prem_races = dndf[~dndf['Race'].isin(free)].groupby('Race').sum().sort_values(by='Count')
```

```python
total_prem = prem_races.sum()[0]
```

```python
prem_races['Percentage'] = [((i / total_prem) * 100).round(2) for i in prem_races['Count']]
```

```python
prem_races = prem_races.reset_index()
```

This next one is just for me, I feel like that graph it makes would be too misleading to share

```python
mixed_bag = pd.merge(how='outer', left=prem_races, right=free_races).sort_values(by='Percentage')
```

```python
mixed_bag['Free'] = mixed_bag['Race'].isin(free) 
```

### Rough Draft

```python
palette = {True: 'g', False: 'purple'}
sns.set(style="dark")
fig, ax = plt.subplots(figsize=(15,5))
plt.title('Heroic Demographic')
plt.xticks(rotation=45,
           horizontalalignment='right',
           fontsize='large'
          )
ax.set_yticklabels([0, '2%', '4%', '6%', '8%', '10%', '12%'])
c = sns.barplot(data=all_races, x='Race', y='Percentage', palette=all_races['Free'].map(palette));
#Can not get the Legend to show colors so, leaving it off.
# fig.savefig('all_races.png')
```

```python
sns.set(style="dark")
fig, ax = plt.subplots()
plt.title('Free Races')
plt.xticks(rotation=45,
           horizontalalignment='right',
           fontsize='large'
          )
ax.set_yticklabels([0, '2.5%', '5%', '7.5%', '10%', '12.5%', '15%', '17.5%'])
sns.barplot(data=free_races, x='Race', y='Percentage', color='g');
# fig.savefig('free_races.png')
```

```python
sns.set(style="dark")
fig, ax = plt.subplots(figsize=(10,5))
plt.title('Premium Races')
plt.xticks(rotation=45,
           horizontalalignment='right',
           fontsize='large'
          )
ax.set_yticklabels([0, '2%', '4%', '6%', '8%', '10%', '12%', '14%', '16%'])
sns.barplot(data=prem_races, x='Race', y='Percentage', color='purple');
# fig.savefig('prem_races.png')
```
